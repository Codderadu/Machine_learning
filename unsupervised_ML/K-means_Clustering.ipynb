```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
X,y=make_blobs(n_samples=1000,centers=3,n_features=2,random_state=23)
```

```python
X.shape
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(
    X,y,test_size=0.33,random_state=42
)
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3,random_state=0,init="k-means++").fit(X_train)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
## manual Process
## Elbow Method to select the k value

wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init="k-means++")
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
#plot elbow curve
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of Clustrers")
plt.ylabel("wcss")
plt.show()
```

```python
kmean=KMeans(n_clusters=3,init="k-means++")
```

```python
y_labels=kmean.fit_predict(X_train)
```

```python
plt.scatter(X_train[:,0],X_train[:,1],c=y_labels)
```

```python
y_test_labels=kmean.predict(X_test)
```

```python
y_test_labels
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_test_labels)
```

```python
## Perfromance metrics
# Silhoutte score
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients=[]
for k in range(2,11):
    kmean=KMeans(n_clusters=k,init="k-means++")
    kmean.fit(X_train)
    score=silhouette_score(X_train,kmean.labels_)
    silhouette_coefficients.append(score)
```

```python
silhouette_coefficients
```

```python
#plot elbow curve
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel("Number of Clustrers")
plt.ylabel("silhouette_coefficients")
plt.show()
```
